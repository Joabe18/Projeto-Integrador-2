# Tratamento de Dados

## Cellphone Data

Alterando cellphone_id com valor zero para cellphone_id = 33

```python
celldata2 = pd.DataFrame.copy(celldata)
```

```python
celldata2.cellphone_id.replace(0,33, inplace=True)
```

Alterando nome das colunas para retirar os espaços

```python
celldata2 = celldata2.rename(columns={'operating system':'operating_system','internal memory':'internal_memory','main camera':'main_camera', \
                          'selfie camera':'selfie_camera','battery size':'battery_size','screen size':'screen_size','release date':'release_date'})
```

```python
celldata2.head()
```

Alterando coluna "release date" que está em formato String para formato data

```python
celldata2['release_date'] = pd.to_datetime(celldata2['release_date'],dayfirst=False)
```

```python
celldata2.head()
```

```python
celldata2.dtypes
```

## Cellusers

```python
cellusers2 = pd.DataFrame.copy(cellusers)
```

```python
cellusers2.head()
```

Tirando os elementos da coluna "gender" de dentro do vetor

```python
cellusers2.gender.replace("['M']",'M', inplace=True)
```

```python
cellusers2.gender.replace("['F']",'F', inplace=True)
```

```python
cellusers2.head()
```
